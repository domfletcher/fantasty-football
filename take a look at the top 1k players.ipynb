```python
with open('top1k.current.json') as data_file:    
    data = json.load(data_file)
```

```python
players = {}
for i in data.keys():
    for j in data[i]:
        if j[u'id'] in players.keys():
            players[j[u'id']] = players[j[u'id']] + 1
        else:
            players[j[u'id']] = 1
```

```python
sorted_players = sorted(players.items(), key=operator.itemgetter(1),reverse=True)
```
